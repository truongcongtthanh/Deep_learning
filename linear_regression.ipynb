```python
import numpy as np
from numpy import genfromtxt
import matplotlib.pyplot as plt

data = genfromtxt('data/data1.csv', dtype = float, delimiter = ',')
m = data.shape[0]
print(type(data))
X = data[:,:-1]
Y = data[:,-1:]
print(X.shape, Y.shape)
X_b = np.c_[np.ones((m,1)), X]
print(X_b)

plt.scatter(X,Y)
plt.xlabel('Din tch nh (x 100$m^2$)')
plt.ylabel('Gi nh (chc lng vng)')
plt.show()
```

### Dng tt c b d liu

```python
def batch_gradient_descent():
    n = 50
    learning_rate = 0.01
# Khi to gi tr tham s thetas
    thetas = np.random.rand(2,1) # tr v mng c kch thc (2,1)
    thetas_path = [thetas]
    print(type(thetas))
    print(type(thetas_path))
    losses = []
    for i in range(n):
        output = X_b.dot(thetas) # Php nhn 2 ma trn (10,2)x(2,1)
        
        loss = (output - Y)**2
        #Cng thc tnh o hm cho loss = 2*X*(o-y)/m (m: l s b d liu  train)
        loss_grd = 2*(output-Y)/m
        gradients = X_b.T.dot(loss_grd)
       # print("lost_grd: ", loss_grd.shape)
        #print("gra: ",gradients.shape)
        # Cp nht tham s
        thetas = thetas - learning_rate*gradients
        thetas_path.append(thetas)
        mean_loss = np.sum(loss)/m
        losses.append(mean_loss)
    return thetas_path, losses
thetas_path, losses = batch_gradient_descent()

plt.scatter(X,Y)
data_Y = X*thetas_path[-1][1] + thetas_path[-1][0]
plt.plot(X, data_Y, color = 'r')
plt.show()

print(thetas_path[-1])
```
